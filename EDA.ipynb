```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
# ✅ اختيار الأعمدة المهمة
df = df[['ProductId', 'UserId', 'ProfileName',
         'HelpfulnessNumerator', 'HelpfulnessDenominator',
         'Score', 'Time', 'Summary', 'Text']]

# ✅ حساب نسبة فائدة المراجعة
df['Helpfulness'] = df['HelpfulnessNumerator'] / df['HelpfulnessDenominator'].replace(0, 1)

# ✅ تحويل الوقت لتاريخ قابل للقراءة
df['ReviewTime'] = pd.to_datetime(df['Time'], unit='s')
df['ReviewYear'] = df['ReviewTime'].dt.year

# ✅ استعراض أول 5 صفوف
df.head()
```

```python
sns.countplot(data=df, x='Score')
plt.title('Distribution of Review Scores')
plt.xlabel('Score')
plt.ylabel('Number of Reviews')
plt.show()
```

```python

df['Review_Length'] = df['Text'].apply(lambda x: len(str(x).split()))
sns.histplot(df['Review_Length'], bins=50)
plt.title('Review Length Distribution')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

```python
df['ReviewYear'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Reviews per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df['ProductId'].value_counts().head(10)
top_products.plot(kind='bar')
plt.title('Top 10 Most Reviewed Products')
plt.xlabel('Product ID')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
df_corr = df[['Score', 'Review_Length']].corr()
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
